### 학습이 끝난 후 테스트데이터를 통해 학습의 결과를 살펴봄

```python
import tensorflow as tf
```

```python
x = tf.placeholder(tf.float32, shape = [None]) # shape = None 은 데이터의 갯수를 알 수 없을 때 사용하는 표현.
y = tf.placeholder(tf.float32, shape = [None])
```

```python
w = tf.Variable(tf.random_normal(shape = [1]), name = 'weight') # 1 * 1의 행렬
```

```python
b = tf.Variable(tf.random_normal(shape = [1]), name = 'bias')
```

```python
hypothesis = x * w + b
```

```python
cost = tf.reduce_mean(tf.square(hypothesis - y))
```

```python
optimizer = tf.train.GradientDescentOptimizer(learning_rate = 0.01)
```

```python
train = optimizer.minimize(cost)
```

```python
sess = tf.Session()
```

```python
sess.run(tf.global_variables_initializer())
```

```python
for i in range(10000) : 
    cost_val, w_val, b_val, _ = sess.run([cost, w, b, train], feed_dict={x:[1, 2, 3], y:[1.5, 2.5, 3.5]})
    if i % 20 == 0 :
        print(cost_val, w_val, b_val) # 학습이 종료되면 cost_val이 최소값이 되어야 함
```

##### 학습이 잘 되었는지 테스트하기

```python
print(sess.run(hypothesis, feed_dict={x:[5]}))
```

```python
print(sess.run(hypothesis, feed_dict={x:[3]}))
```

```python
print(sess.run(hypothesis, feed_dict={x:[1, 3.5]}))
```

```python
sess.close()
```
