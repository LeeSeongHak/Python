파일로부터 데이터를 읽어 x,y로 분리한 후 작업

```python
import tensorflow as tf
import numpy as np
```

```python
data = np.loadtxt('data-01-test-score.csv', delimiter=',', dtype=np.float32)
```

```python
x_data = data[ : , 0:-1]       # -1은 가장 끝 데이터를 의미함.
y_data = data[ : , [-1]]       # [-1]은 가장 마지막 열을 의미함
```

```python
#2차원 매트릭스는 변수 대문자로 씀
X = tf.placeholder(tf.float32, shape=[None, 3])
Y = tf.placeholder(tf.float32, shape=[None, 1])
```

```python
W = tf.Variable(tf.random_normal([3, 1]), name='weight')
b = tf.Variable(tf.random_normal([1]), name='bias')
```

```python
hypothesis = tf.matmul(X, W) + b
```

```python
cost = tf.reduce_mean(tf.square(hypothesis - Y))
```

```python
train = tf.train.GradientDescentOptimizer(learning_rate=1e-5).minimize(cost)
```

```python
sess = tf.Session()
```

```python
init = tf.global_variables_initializer()
```

```python
sess.run(init)
```

```python
for step in range(100000):
    _, cost_val, W_val, b_val = sess.run([train, cost, W, b], feed_dict={X:x_data, Y:y_data})  
    if step % 200 == 0:
        print(cost_val)
    
```
