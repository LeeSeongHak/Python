### 0, 1로 분류하는 logistic regression의 첫 번째 예제
- sigmoid() 사용 : 0~1 사이의 값만 출력하는 함수

```python
import tensorflow as tf
```

```python
x_data = [[1, 2], [2, 3], [3, 1], [4, 3], [5, 3], [6, 2]] # 행의 갯수는 전체 데이터의 갯수이기 때문에 행보다는 열의 갯수가 더 중요.
y_data = [[0],[0],[0],[1],[1],[1]] # 합격은 1, 불합격은 0이 뜨도록.
```

```python
x = tf.placeholder(tf.float32, shape=[None, 2])
Y = tf.placeholder(tf.float32, shape=[None, 1])
```

```python
w = tf.Variable(tf.truncated_normal([2, 1]), name='weight')
b = tf.Variable(tf.truncated_normal([1]), name='bias')
```

```python
hypothesis = tf.sigmoid(tf.matmul(x, w) + b)
```

```python
cost = - tf.reduce_mean(Y * tf.log(hypothesis) + (1 - Y) * tf.log(1 - hypothesis))
```

```python
train = tf.train.GradientDescentOptimizer(learning_rate = 0.01).minimize(cost)
```

### 예측값이 > 0.5이면 1로, 아니면 0으로 수정

```python
predicted = tf.cast(hypothesis > 0.5, dtype=tf.float32)
```

##### 정확성 체크 (predicted 값이 label과 얼마나 맞는지 평균을 구하여 얼마나 정확한지 확률을 구함)

```python
accuracy = tf.reduce_mean(tf.cast(tf.equal(predicted , Y), dtype = tf.float32))
```

```python
with tf.Session() as sess :
    sess.run(tf.global_variables_initializer())
    
    for step in range(100000) :
        cost_val, _ = sess.run([cost, train], feed_dict={x:x_data, Y:y_data})
        
        if step % 200 == 0 :
            print(step, cost_val)
            
    hy_val, pre_val, acc_val = sess.run([hypothesis, predicted, accuracy], feed_dict={x:x_data, Y:y_data})
    print('\nHypothesis : ', hy_val, '\nPredicted :', pre_val, '\nAccuracy : ', acc_val)
```
