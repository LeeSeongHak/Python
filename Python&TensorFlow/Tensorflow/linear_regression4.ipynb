### 파일로부터 데이터 읽은 후 작업하기

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt('simple_data.csv', delimiter = ',', dtype = np.float32 )
```

```python
xdata = data[:, 0]
ydata = data[:, 1]

print(xdata.ndim)
print(xdata.shape)
```

```python
plt.plot(xdata, ydata, 'g', alpha = 0.3)
```

```python
plt.show()
```

```python
x = tf.placeholder(tf.float32, shape = [None]) # shape = None 은 데이터의 갯수를 알 수 없을 때 사용하는 표현.
y = tf.placeholder(tf.float32, shape = [None])
```

```python
w = tf.Variable(tf.random_normal(shape = [1]), name = 'weight') # 1 * 1의 행렬
```

```python
b = tf.Variable(tf.random_normal(shape = [1]), name = 'bias')
```

```python
hypothesis = x * w + b
```

```python
cost = tf.reduce_mean(tf.square(hypothesis - y))
```

```python
optimizer = tf.train.GradientDescentOptimizer(learning_rate = 0.005)
```

```python
train = optimizer.minimize(cost)
```

```python
sess = tf.Session()
```

```python
sess.run(tf.global_variables_initializer())
```

```python
for i in range(10000) : 
    cost_val, w_val, b_val, _ = sess.run([cost, w, b, train], feed_dict={x:xdata, y:ydata})
    if i % 20 == 0 :
        print(cost_val, w_val, b_val)
```
